```python
import os
from metric_extraction import analyze_reviews, TEMPLATE

# Game ID to name mapping
GAME_NAMES = {
    "990080": "Vampire Survivors",
    "1086940": "Baldur's Gate 3",
    "2767030": "Lethal Company",
    "570": "Dota 2",
    "1326470": "Sons Of The Forest",
    "1091500": "Cyberpunk 2077",
    "2669320": "Palworld",
    "2933620": "Enshrouded",
    "271590": "Grand Theft Auto V",
    "1144200": "Ready or Not",
    "2694490": "Manor Lords",
    "2527500": "Helldivers 2",
    "2995920": "Pacific Drive",
    "578080": "PUBG: BATTLEGROUNDS",
    "1174180": "Red Dead Redemption 2",
    "730": "Counter-Strike 2",
    "2183900": "Remnant II",
    "1245620": "Elden Ring",
    "1172470": "Apex Legends",
    "236390": "War Thunder",
}

# Get all processed review files
processed_dir = "../data/processed"
review_files = [f for f in os.listdir(processed_dir) if f.endswith("_processed.txt")]

# Process each file
for file_name in review_files:
    # Extract game ID from filename (remove "review_" and "_processed.txt")
    game_id = file_name.replace("review_", "").replace("_processed.txt", "")
    game_name = GAME_NAMES.get(game_id, f"Unknown Game ({game_id})")

    print(f"\nAnalyzing {game_name}...")
    file_path = os.path.join(processed_dir, file_name)

    with open(file_path, "r", encoding="utf-8") as f:
        reviews = f.read().split("\n\n")
        reviews = [r.strip() for r in reviews if r.strip()]

    # Take first 100 reviews as a test (remove slice to process all)
    test_reviews = reviews

    # Run the analysis
    results = analyze_reviews(test_reviews, TEMPLATE)

    # Print the results
    print(f"Analyzed {len(test_reviews)} reviews")
    print("\nAnalysis Results:")
    for category, result in results.items():
        print(f"{category}: {result}")
    print("-" * 50)
```
